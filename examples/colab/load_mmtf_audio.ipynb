## üöÄ **[Step 2] Use the Framework**

### *1. Load Configurations and Imports*

```python
import os
import json
import pandas as pd
from popcorn.utils import readConfigs

# Start the Framework
print("Welcome to 'Popcorn' üçø! Starting the framework for your movie recommendation ...\n")

# Read the configuration file
configs = readConfigs("Popcorn/popcorn/config/config.yml")
# If properly read, print the configurations
if not configs:
    print("Error reading the configuration file!")
```

### *2. Load MMTF-14K Fused with Text (Audio Variant - i-Vector)*

```python
from popcorn.datasets.mmtf14k.helper_audio import loadAudioFusedDF

# Configurations Override (Optional)
configs["datasets"]["multimodal"]["mmtf"]["audio_variant"] = "ivec"

# Load MMTF-14K Audio
dfAudio = loadAudioFusedDF(configs)
print(f"- Loaded audio fused DataFrame {dfAudio.shape} head:")
dfAudio.head()
```

### *2. Load MMTF-14K Fused with Text (Audio Variant - Block-level Features)*

```python
from popcorn.datasets.mmtf14k.helper_audio import loadAudioFusedDF

# Configurations Override (Optional)
configs["datasets"]["multimodal"]["mmtf"]["audio_variant"] = "blf"

# Load MMTF-14K Audio
dfAudio = loadAudioFusedDF(configs)
print(f"- Loaded audio fused DataFrame {dfAudio.shape} head:")
dfAudio.head()
```
