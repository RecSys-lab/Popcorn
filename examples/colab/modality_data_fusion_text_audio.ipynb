## üöÄ **[Step 2] Use the Framework**

### *1. Load Configurations and Imports*

```python
import os
import json
import pandas as pd
from popcorn.utils import readConfigs

# Start the Framework
print("Welcome to 'Popcorn' üçø! Starting the framework for your movie recommendation ...\n")

# Read the configuration file
configs = readConfigs("Popcorn/popcorn/config/config.yml")
# If properly read, print the configurations
if not configs:
    print("Error reading the configuration file!")
```

### *2. Fuse Poison-RAG-Plus and MMTF-14K datasets*

```python
from popcorn.modalities.fuse_audio_text import (
    fuseTextualAudio_PoisonRag_MMTF14K
)

# Override configs
configs["modalities"]["output_path"] = "/content/Output"

fusedDF = fuseTextualAudio_PoisonRag_MMTF14K(configs)
if fusedDF is None:
    print("- [Error] Fusion failed!")

print(f"- Fused CSV files are saved in '/content/Output'")
```
