### *1. Load Configurations and Imports*

```python
import os
import json
import pandas as pd
from popcorn.utils import readConfigs

# Start the Framework
print("Welcome to 'Popcorn' 🍿! Starting the framework for your movie recommendation ...\n")

# Read the configuration file
configs = readConfigs("Popcorn/popcorn/config/config.yml")
# If properly read, print the configurations
if not configs:
    print("Error reading the configuration file!")

# Get common configurations
cfgMovieLens = configs["datasets"]["unimodal"]["movielens"]
downloadPath = cfgMovieLens["download_path"]

# Override (optional)
downloadPath = "/content/MovieLens"
print(f"- Changing download path from '{cfgMovieLens['download_path']}' to: {downloadPath}")
```

### *3-A. Load MovieLens 25M Thumbnail Embeddings of a Given Movie (Dino-V2)*

```python
from popcorn.datasets.ml_thumbnail.helper_embedding import (
    loadMovieLensThumbnailEmbeddings,
    loadAllMovieLensThumbnailEmbeddings
)

# Variables
root = "/content/Thumbnails"
configs["datasets"]["unimodal"]["ml_thumbnail"]["variant"] = 'dino-v2'
configs["datasets"]["unimodal"]["ml_thumbnail"]["download_path"] = root

# Load thumbnail embeddings for a given part
partId = 1
variant = configs["datasets"]["unimodal"]["ml_thumbnail"]["variant"]
embeddingsDF = loadMovieLensThumbnailEmbeddings(partId, variant)

display(embeddingsDF)
```

### *3. Load MovieLens 25M Thumbnail Embeddings of a Given Movie (CLIP)*

```python
from popcorn.datasets.ml_thumbnail.helper_embedding import (
    loadMovieLensThumbnailEmbeddings,
    loadAllMovieLensThumbnailEmbeddings
)

# Variables
root = "/content/Thumbnails"
configs["datasets"]["unimodal"]["ml_thumbnail"]["variant"] = 'clip'
configs["datasets"]["unimodal"]["ml_thumbnail"]["download_path"] = root

# Load thumbnail embeddings for a given part
partId = 1
variant = configs["datasets"]["unimodal"]["ml_thumbnail"]["variant"]
embeddingsDF = loadMovieLensThumbnailEmbeddings(partId, variant)

display(embeddingsDF)
```

### *4. Load All ML-25M Thumbnail Embeddings*

```python
# Load the Movie Thumbnail
allEmbeddingsDF = loadAllMovieLensThumbnailEmbeddings(variant)

display(allEmbeddingsDF)
```
