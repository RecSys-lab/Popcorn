### *1. Load Configurations and Imports*

```python
import os
import json
import pandas as pd
from popcorn.utils import readConfigs

# Start the Framework
print("Welcome to 'Popcorn' üçø! Starting the framework for your movie recommendation ...\n")

# Read the configuration file
configs = readConfigs("Popcorn/popcorn/config/config.yml")
# If properly read, print the configurations
if not configs:
    print("Error reading the configuration file!")
```

### *2. Check overlap between Poison-RAG-Plus and Popcorn datasets*

```python
from popcorn.modalities.helper_overlap import checkOverlap_PoisonRag_Popcorn

overlappedDF = checkOverlap_PoisonRag_Popcorn(configs)
if overlappedDF is None:
    print("- Overlap check failed!")
overlappedDF.head(5)
```

### *3. Check Overlap between Poison-RAG-Plus and MMTF-14K Datasets*

```python
from popcorn.modalities.helper_overlap import checkOverlap_PoisonRag_MMTF14K

overlappedDF = checkOverlap_PoisonRag_MMTF14K(configs)
if overlappedDF is None:
    print("- Overlap check failed!")
overlappedDF.head(5)
```

### *4. Check Overlap between Popcorn and MMTF-14K Datasets*

```python
from popcorn.modalities.helper_overlap import checkOverlap_Popcorn_MMTF14K

overlappedDF = checkOverlap_Popcorn_MMTF14K(configs)
if overlappedDF is None:
    print("- Overlap check failed!")
overlappedDF.head(5)
```

### *5. Check Overlap among All (Poison-RAG-Plus, Popcorn, and MMTF-14K)*

```python
from popcorn.modalities.helper_overlap import checkOverlap_All

overlappedDF = checkOverlap_All(configs)
if overlappedDF is None:
    print("- Overlap check failed!")
overlappedDF.head(5)
```
