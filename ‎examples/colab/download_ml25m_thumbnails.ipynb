## 🚀 **[Step 2] Use the Framework**

### *1. Load Configurations and Imports*

```python
import os
import json
import pandas as pd
from popcorn.utils import readConfigs
from popcorn.datasets.movielens.loader import loadMovieLens
from popcorn.datasets.utils import printTextualDatasetStats

# Start the Framework
print("Welcome to 'Popcorn' 🍿! Starting the framework for your movie recommendation ...\n")

# Read the configuration file
configs = readConfigs("Popcorn/popcorn/config/config.yml")
# If properly read, print the configurations
if not configs:
    print("Error reading the configuration file!")
```

### *1-A. TMDB-API Set*

It fetches movie poster thumbnails using the TMDB API ([link](https://developer.themoviedb.org/docs/getting-started)).

```python
from google.colab import userdata
os.environ["TMDB_TOKEN"] = userdata.get("TMDB_TOKEN")

# Receive it from TMDB
TMDB_API_KEY = os.getenv("TMDB_API_KEY", os.environ["TMDB_TOKEN"])
```

### *2. Download & Load MovieLens 25M*

Only this version of MovieLens has a dataframe containing IMDB and TMBD links.

```python
from popcorn.datasets.movielens.loader import loadMovieLens
from popcorn.datasets.movielens.downloader import downloadMovieLens

# Variables
downloadPath = "/content/MovieLens" # For Google Colab Drive

# Override cofigurations
configs["datasets"]["unimodal"]["movielens"]["version"] = "25m"

# Download MovieLens dataset - 25m
downloadMovieLens('25m', downloadPath)

# Load the dataset
itemsDF, usersDF, ratingsDF, linksDF = loadMovieLens(configs)
```

### *3. List of Movies to Download their Thumbnails - Dummy Dataset*

```python
from popcorn.pipelines.thumbnail_fetch.core import getMovieList
from popcorn.pipelines.thumbnail_fetch.downloader import downloadThumbnails

# Variables
datasetVariant = "dummy"

# Override configurations
configs["pipelines"]["thumbnail_fetch"]["tmdb_api_key"] = TMDB_API_KEY
configs["pipelines"]["thumbnail_fetch"]["dataset_variant"] = datasetVariant
configs["pipelines"]["thumbnail_fetch"]["download_path"] = "/content/Thumbnails"

# Load MovieLens
movieList = getMovieList(datasetVariant, configs)

# Download
downloadThumbnails(movieList, linksDF, configs)
```
